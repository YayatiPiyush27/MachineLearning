```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

''' download iris.csv
https://gist.github.com/Thanatoz-1/9e7fdfb8189f0cdf5d73a494e4a6392a/raw/aaecbd14aeaa468cd749528f291aa8a30c2ea09e/iris_dataset.csv
'''

#load iris.csv into googleColab

iris=pd.read_csv("iris_dataset.csv")
```

```python
#How many data-points/enries/rows and features/dimentions/columns

print(iris.shape)
print(iris.columns)
```

```python
#How many data points for each CLASS ??
#how many flowers of each spcies??

iris["target"].value_counts()
```

# 2-D Scatter Plot

```python
iris.plot(kind='scatter',x='sepal length (cm)',y='sepal width (cm)');
plt.show()
#cant make sense out of it
#which flower which species
```

# 2-D Scatter Plot With Color Coding

```python
#we will use seaborn
sns.set_style("whitegrid");
sns.FacetGrid(iris,hue="target",height=4.5) \
.map(plt.scatter,"sepal length (cm)","sepal width (cm)") \
.add_legend();
plt.show();
```

#Observations:
1. We can lineraly Sepearte Blue VS Others
2. Orange and Green are Not Linearly Seperable
